## Step 1: Load Processed Data

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

with open('../data/processed_data.pkl', 'rb') as f:
    data = pickle.load(f)

X_train, X_test = data['X_train'], data['X_test']
y_train, y_test = data['y_train'], data['y_test']
X_train_scaled = data['X_train_scaled']
X_test_scaled = data['X_test_scaled']
le_type = data['le_type']
type_classes = le_type.classes_

print(f'Training samples: {len(X_train)}')
print(f'Test samples: {len(X_test)}')
```

## Step 2: Compare Different Kernels

```python
kernels = ['linear', 'rbf']
kernel_results = {}

for kernel in kernels:
    svc = SVC(kernel=kernel, random_state=42)
    svc.fit(X_train_scaled, y_train)
    y_pred = svc.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred)
    kernel_results[kernel] = acc
    print(f'{kernel.capitalize()} kernel - Accuracy: {acc:.4f}')
```

## Step 3: Train Final Model with Best Kernel

```python
best_kernel = max(kernel_results, key=kernel_results.get)
print(f'Best kernel: {best_kernel} (accuracy: {kernel_results[best_kernel]:.4f})')

svm = SVC(kernel=best_kernel, random_state=42)
svm.fit(X_train_scaled, y_train)
y_pred = svm.predict(X_test_scaled)
print('Model trained successfully')
```

## Step 4: Evaluate Model

```python
print(f'Accuracy: {accuracy_score(y_test, y_pred):.4f}')
print(f'Precision: {precision_score(y_test, y_pred, average="weighted", zero_division=0):.4f}')
print(f'Recall: {recall_score(y_test, y_pred, average="weighted", zero_division=0):.4f}')
print(f'F1-Score: {f1_score(y_test, y_pred, average="weighted", zero_division=0):.4f}')
```

```python
print('Classification Report:')
print(classification_report(y_test, y_pred, target_names=type_classes, zero_division=0))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
plt.imshow(cm, interpolation='nearest', cmap='Blues')
plt.title('SVM Confusion Matrix', fontsize=14)
plt.colorbar()
tick_marks = np.arange(len(type_classes))
plt.xticks(tick_marks, type_classes, rotation=45, ha='right')
plt.yticks(tick_marks, type_classes)
for i in range(len(type_classes)):
    for j in range(len(type_classes)):
        color = 'red' if cm[i, j] > cm.max() / 2 else 'black'
        plt.text(j, i, str(cm[i, j]), ha='center', va='center', color=color)
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.savefig('../reports/svm_confusion_matrix.png', dpi=150)
plt.show()
```

## Step 5: Save Model

```python
with open('../data/svm_model.pkl', 'wb') as f:
    pickle.dump(svm, f)
print('Model saved to: ../data/svm_model.pkl')
```
