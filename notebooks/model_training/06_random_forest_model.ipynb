## Step 1: Load Processed Data

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

with open('../data/processed_data.pkl', 'rb') as f:
    data = pickle.load(f)

X_train, X_test = data['X_train'], data['X_test']
y_train, y_test = data['y_train'], data['y_test']
le_type = data['le_type']
type_classes = le_type.classes_
feature_cols = data['feature_cols']

print(f'Training samples: {len(X_train)}')
print(f'Test samples: {len(X_test)}')
```

## Step 2: Implement Random Forest

```python
rf = RandomForestClassifier(n_estimators=100, max_depth=7, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print('Model trained successfully')
```

## Step 3: Evaluate Model

```python
print(f'Accuracy: {accuracy_score(y_test, y_pred):.4f}')
print(f'Precision: {precision_score(y_test, y_pred, average="weighted", zero_division=0):.4f}')
print(f'Recall: {recall_score(y_test, y_pred, average="weighted", zero_division=0):.4f}')
print(f'F1-Score: {f1_score(y_test, y_pred, average="weighted", zero_division=0):.4f}')
```

```python
print('Classification Report:')
print(classification_report(y_test, y_pred, target_names=type_classes, zero_division=0))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
plt.imshow(cm, interpolation='nearest', cmap='Blues')
plt.title('Random Forest Confusion Matrix', fontsize=14)
plt.colorbar()
tick_marks = np.arange(len(type_classes))
plt.xticks(tick_marks, type_classes, rotation=45, ha='right')
plt.yticks(tick_marks, type_classes)
for i in range(len(type_classes)):
    for j in range(len(type_classes)):
        color = 'red' if cm[i, j] > cm.max() / 2 else 'black'
        plt.text(j, i, str(cm[i, j]), ha='center', va='center', color=color)
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.savefig('../reports/rf_confusion_matrix.png', dpi=150)
plt.show()
```

## Step 4: Feature Importance

```python
importances = rf.feature_importances_
feature_names = [c.replace('_enc', '').replace('_', ' ').title() for c in feature_cols]

plt.figure(figsize=(8, 6))
plt.barh(feature_names, importances, color='forestgreen')
plt.title('Random Forest Feature Importance', fontsize=14)
plt.xlabel('Importance', fontsize=12)
plt.tight_layout()
plt.savefig('../reports/rf_feature_importance.png', dpi=150)
plt.show()

for name, imp in zip(feature_names, importances):
    print(f'{name}: {imp:.4f}')
```

## Step 5: Save Model

```python
with open('../data/random_forest_model.pkl', 'wb') as f:
    pickle.dump(rf, f)
print('Model saved to: ../data/random_forest_model.pkl')
```
