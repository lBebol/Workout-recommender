## Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

plt.rcParams['figure.figsize'] = (10, 6)
plt.style.use('seaborn-v0_8-whitegrid')
```

## Step 2: Load Dataset

```python
df = pd.read_csv('../data/workout_dataset.csv')
print(f'Dataset Shape: {df.shape}')
print(f'Total Exercises: {len(df)}')
print(f'Number of Features: {len(df.columns)}')
```

## Step 3: Initial Inspection

```python
print('First 5 rows:')
df.head()
```

```python
print('Data Types:')
df.dtypes
```

```python
print('Missing Values:')
df.isnull().sum()
```

## Step 4: Visualizations

```python
os.makedirs('../reports', exist_ok=True)

# Plot 1: Distribution by Target Muscle
plt.figure(figsize=(10, 6))
counts = df['target_muscles'].value_counts().sort_values(ascending=False)
plt.bar(counts.index, counts.values, color='steelblue')
plt.title('Distribution of Exercises by Target Muscle', fontsize=14)
plt.xlabel('Target Muscle', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../reports/muscle_distribution.png', dpi=150)
plt.show()
```

```python
# Plot 2: Difficulty Distribution
plt.figure(figsize=(8, 6))
diff_counts = df['difficulty'].value_counts()
colors = {'beginner': '#2ecc71', 'intermediate': '#f39c12', 'advanced': '#e74c3c'}
wedges, texts, autotexts = plt.pie(
    diff_counts.values,
    labels=diff_counts.index,
    colors=[colors.get(d, '#3498db') for d in diff_counts.index],
    autopct='%1.1f%%',
    startangle=90,
    textprops={'fontsize': 12}
)
for t in autotexts:
    t.set_fontsize(11)
    t.set_fontweight('bold')
plt.title('Exercise Difficulty Distribution', fontsize=14)
plt.tight_layout()
plt.savefig('../reports/difficulty_distribution.png', dpi=150)
plt.show()
```

```python
# Plot 3: Exercise Type Distribution
plt.figure(figsize=(10, 6))
type_counts = df['type'].value_counts().sort_values(ascending=False)
plt.bar(type_counts.index, type_counts.values, color='coral')
plt.title('Distribution by Exercise Type', fontsize=14)
plt.xlabel('Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../reports/type_distribution.png', dpi=150)
plt.show()
```

```python
# Plot 4: Equipment Distribution
plt.figure(figsize=(10, 6))
equip_counts = df['equipment'].value_counts().sort_values(ascending=False)
plt.bar(equip_counts.index, equip_counts.values, color='mediumseagreen')
plt.title('Distribution by Equipment', fontsize=14)
plt.xlabel('Equipment Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../reports/equipment_distribution.png', dpi=150)
plt.show()
```

```python
# Plot 5: Heatmap (Muscle vs Difficulty)
plt.figure(figsize=(10, 6))
cross_tab = pd.crosstab(df['target_muscles'], df['difficulty'])
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='YlOrRd', linewidths=0.5,
            cbar_kws={'label': 'Number of Exercises'})
plt.title('Muscle Groups by Difficulty', fontsize=14)
plt.xlabel('Difficulty', fontsize=12)
plt.ylabel('Target Muscle', fontsize=12)
plt.tight_layout()
plt.savefig('../reports/heatmap.png', dpi=150)
plt.show()
```
